```python
lst_cate_column = []
for column in df.columns:
    if (df[column].dtype == 'O'):
        lst_cate_column.append(column)

rows = 3
columns = 5
count = 0
fig, axs = plt.subplots(rows, columns, figsize=(40,40))
for i in range(rows):
    for j in range(columns):
        sns.barplot(ax=axs[i, j], data=df.groupby(lst_cate_column[count]
                                                   )['cost'].mean().sort_values().to_frame().reset_index()
                    , x=lst_cate_column[count], y='cost')
        axs[i, j].tick_params(axis='x', rotation=90)
        count += 1
```

```python
df = df.drop(columns=['food_category', 'food_department', 'food_family', 'gender', 'marital_status',
                         'brand_name', 'avg. yearly_income', 'education', 'member_card', 'houseowner', 
                          'sales_country', 'occupation'])

```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
import time

```

```python
encoder = OrdinalEncoder()
final_data = encoder.fit_transform(df.drop(columns='cost'))
```

```python
X = final_data
y = df['cost']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = DecisionTreeRegressor()
t0 = time.time()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
t1 = time.time() - t0
print('r2 score: ', r2_score(y_test, y_pred))
print('time : ', t1)
```

```python
model = DecisionTreeRegressor(random_state=44)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```
